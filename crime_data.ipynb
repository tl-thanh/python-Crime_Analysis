```python
crime_df_group = crime_df.groupby(["Primary Type","Year"]).count().reset_index()
crime_df_group
```

```python
crime_df_group = crime_df_group.rename(columns={"Primary Type":"Type"})
crime_df_group
```

```python
type_list = crime_df_group["Type"].drop_duplicates()

pivot_df = crime_df_group.pivot(index="Year",columns='Type',values="Location").fillna(0)
# colors = ['red','lightgreen','blue', 'green', 'orange', 'pink', 'lightblue', 'maroon', 'darksalmon', 'sandybrown', 
#           'darkkhaki', 'olivedrab', 'seagreen', 'darkcyan', 'navy', 'darkorchid', 'plum', 'lightcoral', 'cadetblue', 
#           'dodgerblue', 'rebeccapurple', 'darkviolet', 'fuchsia', 'deeppink', 'crimson', 'indianred', 'darkred', 
#           'aqua', 'aquamarine', 'darkslateblue',"darkblue","darkgreen","purple","yellow"]
colors = sns.light_palette((210, 90, 60), input="husl", reverse=True)
pivot_df.loc[:,type_list].plot.bar(stacked=True,color=colors,figsize=(20,10),fontsize=16)
plt.title("Crimes of Chicago by Type",fontsize=30)
plt.ylabel("Number of crimes",fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.legend(loc="best",fontsize=9.75, title="Crime Type",mode="Expanded", labelspacing=0.5,bbox_to_anchor=(1.0, 1.0))

# Save the Figure
plt.savefig("crimes_type.png")

#show the figure
plt.show()
```

```python
# read parking traffic violation csv data file
parking_pd = pd.read_csv("parking_tickets.csv")
```

```python
columns = ["issue_date","violation_code","violation_description","zipcode","fine_level1_amount","fine_level2_amount"]
parking_df = pd.DataFrame(parking_pd, columns = columns)
parking_df.head()
```

```python
len(parking_df)
```

```python
import datetime
parking_df["issue_date"] = pd.to_datetime(parking_df["issue_date"])
```

```python
parking_df["year"] = parking_df["issue_date"].dt.year
```

```python
parking_df.tail(100)
```

```python
parking_df_group = parking_df.groupby(["violation_description","year"]).count().reset_index()
parking_df_group
```

```python
violation_list = parking_df_group["violation_description"].drop_duplicates()

pivot_df1 = parking_df_group.pivot(index="year",columns="violation_description",values="fine_level1_amount").fillna(0)
# colors = ['red','lightgreen','blue', 'green', 'orange', 'pink', 'lightblue', 'maroon', 'darksalmon', 'sandybrown', 
#           'darkkhaki', 'olivedrab', 'seagreen', 'darkcyan', 'navy', 'darkorchid', 'plum', 'lightcoral', 'cadetblue', 
#           'dodgerblue', 'rebeccapurple', 'darkviolet', 'fuchsia', 'deeppink', 'crimson', 'indianred', 'darkred', 
#           'aqua', 'aquamarine', 'darkslateblue',"darkblue","darkgreen","purple","yellow"]
colors = sns.light_palette((210, 90, 60), input="husl", reverse=True)
pivot_df1.loc[:,violation_list].plot.bar(stacked=True,color=colors,figsize=(20,10),fontsize=16,legend=None)
plt.title("Chicago Traffic Violation",fontsize=30)
plt.ylabel("Number of Traffic Violations",fontsize=20)
plt.xlabel("Year",fontsize=20)
# plt.legend(loc="upper center",fontsize=9.75, title="Traffic Violation Type",mode="Expanded", labelspacing=0.5,bbox_to_anchor=(0.5, -0.05))

# Save the Figure
plt.savefig("traffic_violation.png")

#show the figure
plt.show()
```
